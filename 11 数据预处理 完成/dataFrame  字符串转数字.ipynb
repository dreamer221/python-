```python
import pandas as pd
```

# 第一种方法

```python
path = 'iris.data'  # 数据文件路径
data = pd.read_csv(path, header=None)
data
```

```python
y = pd.Categorical(data[4])
print("y:", y)

y = pd.Categorical(data[4]).codes
print("\ny转换后:", y)
```

```python
data.loc[:, 4] = y
data
```

# 第二种方法

```python
path = 'iris.data'  # 数据文件路径
data = pd.read_csv(path, header=None)
data
```

```python
y = data.loc[:, 4].map({'Iris-setosa': 0, 
                        'Iris-versicolor': 1, 
                        'Iris-virginica':2})
data.loc[:, 4]  = y
data
```
