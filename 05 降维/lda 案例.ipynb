```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, decomposition, discriminant_analysis
```

```python
x1 = np.random.normal(1, 0.1, (25, 3))
x2 = np.random.normal(2, 0.1, (25, 3))
x3 = np.random.normal(3, 0.1, (25, 3))
x4 = np.random.normal(4, 0.1, (25, 3))
x = np.vstack((x1, x2, x3, x4))
y = ['r'] * 25 + ['g'] * 25 + ['b'] * 25 + ['y'] * 25

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
for i in range(len(x)):
    ax.scatter(x[i, 0], x[i, 1], x[i, 2], color=y[i])
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
fig.suptitle("Iris After LDA")
plt.show()
```

```python
lda = discriminant_analysis.LinearDiscriminantAnalysis(n_components=2)
lda.fit(x, y)
print("\n权重系数:\n", lda.coef_)
print("\n常数项:", lda.intercept_)
converted_X = np.dot(x, np.transpose(lda.coef_)) + lda.intercept_
converted_X.shape
```

```python
lda = discriminant_analysis.LinearDiscriminantAnalysis(n_components=1)
lda.fit(x, y)
print("\n权重系数:\n", lda.coef_)
print("\n常数项:", lda.intercept_)
converted_X = np.dot(x, np.transpose(lda.coef_)) + lda.intercept_
converted_X.shape
```
