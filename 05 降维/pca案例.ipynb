```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, decomposition
```

```python
x1 = np.random.normal(1, 0.1, (25, 3))
x2 = np.random.normal(2, 0.1, (25, 3))
x3 = np.random.normal(3, 0.1, (25, 3))
x4 = np.random.normal(4, 0.1, (25, 3))
x = np.vstack((x1, x2, x3, x4))

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x[:, 0], x[:, 1], x[:, 2])
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
fig.suptitle("Iris After LDA")
plt.show()
```

```python
pca = decomposition.PCA(n_components=2)  # 降维2维
pca.fit(x)
x_2d = pca.transform(x)

plt.scatter(x_2d[:, 0], x_2d[:, 1])
plt.show()
```

```python
pca = decomposition.PCA(n_components=1)  # 降维1维
pca.fit(x)
x_1d = pca.transform(x)
plt.scatter(x_1d, [0] * len(x_1d) )
plt.show()
```

```python
X = np.array([[2.8, 3, 3], [2, 2.2, 2], [1, 1, 1.1], [0, 0.2, -0.3]])
pca = decomposition.PCA(n_components=3)
pca.fit(X)  
print(pca.explained_variance_ratio_) 
pca.transform(X)
```
